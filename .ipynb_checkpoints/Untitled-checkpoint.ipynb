```python
TOI_RSS_URL = "https://timesofindia.indiatimes.com/rssfeedstopstories.cms"  # main feed

def fetch_toi_headlines(max_items: int = 15):
    """
    Tool: Fetch headlines from Times of India RSS.
    Returns a list of dicts: {title, link, summary}
    """
    feed = feedparser.parse(TOI_RSS_URL)
    articles = []
    for entry in feed.entries[:max_items]:
        articles.append({
            "title": entry.get("title", "").strip(),
            "link": entry.get("link", "").strip(),
            "summary": entry.get("summary", "").strip()
        })
    return articles
```
